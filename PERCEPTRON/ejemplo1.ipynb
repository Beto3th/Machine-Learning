```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
class Perceptron:
    def __init__(self, learning_rate=0.01, epochs=100):
        self.learning_rate = learning_rate # Que tan rapido aprende
        self.epochs = epochs # Cuantas veces ve los datos


    def fit(self, X, y):
        # Inicializar los pesos en cero
        self.weights = np.zeros(X.shape[1])
        # inicializar el sesgo
        self.bias = 0

        # Entrenamiento del perceptor
        for _ in range(self.epochs):
            # Iteración de datos
            for i in range(X.shape[0]):
                # Calculo de el producto punto entre los pesos y las entradas más el bias
                y_pred = np.dot(X[i], self.weights) + self.bias
                # Aplicar función de activación
                y_pred = 1 if y_pred >= 0 else 0
                # Actualizar pesos de las entradas 
                # Si hay un error en la predicción
                if y_pred != y[i]:
                    update = self.learning_rate * (y[i] - y_pred)
                    self.weights += update * X[i]
                    self.bias += update

    def predict(self, X):
        # Predecir las etiquetas de los datos que no conoce
        y_pred = np.dot(x, self.weights) + self.bias
        # Hace la funcion de activacion para todo el arreglo de y pred
        return np.where(y_pred >= 0, 1,0)
```
