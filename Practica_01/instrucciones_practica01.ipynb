## Dataset
El conjunto de datos del Titanic es un conjunto de datos clásico utilizado para la enseñanza y la práctica en ciencia de datos y machine learning. Contiene información sobre los pasajeros a bordo del Titanic, incluidos detalles como su edad, sexo, clase socioeconómica, tarifa del boleto, etc. Las columnas incluídas en este conjunto de datos son:

1. Survived (Sobrevivido): Esta columna indica si el pasajero sobrevivió al hundimiento del Titanic o no. Es una variable binaria donde 1 representa que el pasajero sobrevivió y 0 representa que no.
2. Pclass (Clase): Esta columna indica la clase socioeconómica del pasajero. Puede tomar uno de tres valores: 1 para primera clase, 2 para segunda clase y 3 para tercera clase.
3. Name (Nombre): Esta columna contiene el nombre del pasajero.
4. Sex (Género): Esta columna indica el género del pasajero. Puede tomar dos valores: "male" para hombres y "female" para mujeres.
5. Age (Edad): Esta columna indica la edad del pasajero en años.
6. SibSp (Número de hermanos/cónyuges a bordo): Esta columna indica el número de hermanos o cónyuges que el pasajero tenía a bordo del Titanic.
7. Parch (Número de padres/hijos a bordo): Esta columna indica el número de padres o hijos que el pasajero tenía a bordo del Titanic.
8. Ticket (Boleto): Esta columna contiene el número de boleto del pasajero.
9. Fare (Tarifa): Esta columna indica la tarifa pagada por el pasajero.
10. Cabin (Cabina): Esta columna contiene el número de cabina del pasajero, aunque muchos valores están ausentes.
11. Embarked (Puerto de embarque): Esta columna indica el puerto de embarque del pasajero. Puede tomar uno de tres valores: "C" para Cherbourg, "Q" para Queenstown y "S" para Southampton.

# Ejercicio 1

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
datos = pd.read_csv("titanic.csv")
```

```python
survived_0 = datos[datos["Survived"] == 0]
survived_1 = datos[datos["Survived"] == 1]
```

```python
plt.figure(figsize=(8,6))
plt.bar(["Sobrevivientes", "Fallecidos"], [len(survived_1), len(survived_0)], color=["skyblue", "pink"])
plt.xlabel("Tripulación")
plt.ylabel("Cantidad")
plt.title("Sobrevivientes y Fallecidos")
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.pie([len(survived_1), len(survived_0)], labels=["Sobrevivientes", "Fallecidos"], autopct='%1.1f%%', startangle=90)
plt.title("Sobrevivientes y Fallecidos")
plt.axis("equal")
plt.show()
```

# Ejercicio 2

```python
clase_1 = datos[datos["Pclass"] == 1]
clase_2 = datos[datos["Pclass"] == 2]
clase_3 = datos[datos["Pclass"] == 3]
```

```python
plt.figure(figsize=(12, 8))
plt.hist(clase_1["Pclass"], alpha=0.5, label="Clase 1")
plt.hist(clase_2["Pclass"], alpha=0.5, label="Clase 2")
plt.hist(clase_3["Pclass"], alpha=0.5, label="Clase 3")
plt.xlabel('Clase')
plt.ylabel('Frecuencia')
plt.title('Distribución de Pclass')
plt.legend()
plt.grid(axis='y')
plt.show()
```

```python
survived_counts = datos.groupby(['Pclass', 'Survived']).size().unstack()

# Obtener las clases de pasajeros (Pclass)
pclasses = survived_counts.index

# Obtener los recuentos de supervivencia y no supervivencia para cada clase de pasajero
survived = survived_counts[1]
not_survived = survived_counts[0]
```

```python
bar_width = 0.35
index = range(len(pclasses))
plt.bar(index, survived, bar_width, label='Sobreviviente', color='pink')
plt.bar(index, not_survived, bar_width, label='No sobreviviente', color='skyblue', bottom=survived)

# Configurar el título y etiquetas de los ejes
plt.xlabel('Clases')
plt.ylabel('Cantidad')
plt.title('Sobrevivientes por clase')
plt.xticks(index, pclasses)
plt.legend()

# Mostrar la gráfica
plt.show()
```
